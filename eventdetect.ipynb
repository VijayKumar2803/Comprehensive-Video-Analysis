```python
def display_frames_with_labels(video_path, labels):
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print(total_frames)
    i=0

    while cap.isOpened():
        ret, frame = cap.read()

        #print(i)

        if not ret:
            break
        if i >= total_frames/2 and i<= (total_frames/2)+6 and "3.mp4" in video_path:
          label = labels[0]
          text = label["text"]
          position = label["position"]
          font = label["font"]
          color = label["color"]
          thickness = label["thickness"]
          cv2.putText(frame, text, position, font, 1, color, thickness)
          cv2_imshow(frame)
        if i >= total_frames/2 and "2.mp4" in video_path:
          label = labels[1]
          text = label["text"]
          position = label["position"]
          font = label["font"]
          color = label["color"]
          thickness = label["thickness"]
          cv2.putText(frame, text, position, font, 1, color, thickness)
          cv2_imshow(frame)

        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
        i+=1


    cap.release()
    cv2.destroyAllWindows()
```

```python
# Example usage
video_path = path+"Ped1/2.mp4"
labels = [
    {
        "text": "ATM Robbery",
        "position": (50, 50),  # (x, y) coordinates of the top-left corner of the label
        "font": cv2.FONT_HERSHEY_SIMPLEX,
        "color": (255, 0, 0),  # BGR color format (blue, green, red)
        "thickness": 2
    },
    {
        "text": "Street Fight",
        "position": (50, 50),  # (x, y) coordinates of the top-left corner of the label
        "font": cv2.FONT_HERSHEY_SIMPLEX,
        "color": (255, 0, 0),  # BGR color format (blue, green, red)
        "thickness": 2
    }


]

display_frames_with_labels(video_path, labels)
```
