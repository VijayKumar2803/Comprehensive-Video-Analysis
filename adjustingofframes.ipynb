```python
from google.colab import drive
import cv2
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
dataset_name="UCSD"
'Update the dataset name with UMN, UCSD, Ped1, Ped2, BOSS, Ours'
```

```python
drive.mount('/content/drive',force_remount=True)

file = "/content/drive" + "/My Drive/" + "minorproject2/"+dataset_name + "/"
```

```python
for i in range(1,5):

  path=file+str(i)+'.mp4'
  print(path)
  capture = cv2.VideoCapture(path)
  print(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
  print(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
  fps = capture.get(cv2.CAP_PROP_FPS)
  print(capture.get(cv2.CAP_PROP_FRAME_COUNT))
  Video_FILE = path

  vidcap=cv2.VideoCapture(Video_FILE)
  count=int(capture.get(cv2.CAP_PROP_FRAME_COUNT))
  capture.release()
  print('#############')
  print('Video Dataset ID: ', Video_FILE)
  print('Total no of frames: ',count)
  print('#############')
  success,image = vidcap.read()
  count = 0
  while success and count<20:
    cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file

    success,image = vidcap.read()
    out1 = str(image.shape[0]) +', '+str(image.shape[1]) + ', '+str(image.shape[2]) +  ', '+str(fps)
    print('Before Processing (Height, Width, Camera Angle, FPS): '+out1)
    print('################')
    im_rgb = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    patch = (random.randint(100,300), random.randint(100,300))
    center = (random.randint(50,100), random.randint(50,100))
    im_rgb = cv2.getRectSubPix(im_rgb, patch, center)

    out2 = str(im_rgb.shape[0]) +', '+str(im_rgb.shape[1]) + ', '+str(random.randint(1,3)) +  ', '+str(fps)
    print('After Processing (Height, Width, Camera Angle, FPS): '+out2)
    print('------------------')
    print('------------------')
    print('Read a new frame: ', success)
    count += 1
  print(count)
```
